### 3. Limpieza de Datos

Verificar y manejar valores nulos

```python
# Verificar si hay valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Seleccionar sólo las columas numéricas
numerical_columns = df.select_dtypes(include=["number"]).columns

# Reemplazar valores nulos por la mediana en las columnas numéricas
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())

# Verificar si se han reemplazo los valores nulos
print("\nDespués de reemplazar valores nulos:")
print(df.isnull().sum())
```

**Pregunta:** ¿Cuál es el porcentaje de datos faltantes en cada columna?

```python
# Calcular el porcentaje de datos faltantes por columna
missing_data_percentage = df.isnull().mean() * 100
print("\nPorcentaje de datos faltantes por columna:")
print(missing_data_percentage)
```

#### Conclusión Intermedia: Limpieza de Datos  

Durante la etapa de **verificación y manejo de valores nulos**, se evaluaron los datos en busca de valores faltantes en cada columna. Al calcular el porcentaje de datos nulos, se determinó que **todas las columnas tienen un 0% de datos faltantes**.  

Como resultado, no fue necesario realizar imputaciones o reemplazos, ya que el conjunto de datos está completamente limpio en este aspecto. Esta verificación garantiza que el análisis posterior no estará afectado por valores ausentes.  

### 4. Análisis Exploratorio

Calcular métricas para las columnas de calificaciones

```python
# Selección de notas por asignatura
df[["Nota_Matemáticas", "Nota_Lenguaje", "Nota_Ciencias"]]
```

```python
# Calcular el promedio por estudiante (promedio de las tres asignaturas)
df["Promedio"] = df[["Nota_Matemáticas", "Nota_Lenguaje", "Nota_Ciencias"]].mean(axis=1)

print("\nPromedio por estudiante (primeros 5):")
df[["Estudiante_ID", "Promedio"]].head()
```

Determinar el porcentaje de estudiantes con calificación mayor o igual a 60

```python
# Calcular el porcentaje de estudiantes con una calificación mayor o igual a 60 en las tres asignaturas
porcentaje_60_matematicas = (df["Nota_Matemáticas"] >= 60).mean() * 100
porcentaje_60_lenguaje = (df["Nota_Lenguaje"] >= 60).mean() * 100
porcentaje_60_ciencias = (df["Nota_Ciencias"] >= 60).mean() * 100

print("\nPorcentaje de estudiantes con calificación >= 60 en cada asignatura:")
print(f"Matemáticas: {porcentaje_60_matematicas:.2f}%")
print(f"Lenguaje: {porcentaje_60_lenguaje:.2f}%")
print(f"Ciencias: {porcentaje_60_ciencias:.2f}%")
```

**Pregunta:** ¿Qué comuna tiene el promedio más alto en matemáticas?

```python
# Agrupar los datos por comuna y calcular el promedio de matemáticas por comuna
promedio_comuna = df.groupby("Comuna")["Nota_Matemáticas"].mean()

print("\nComuna con el promedio más alto en matemáticas:")
promedio_comuna.nlargest(1)
```

### 6. Preprocesamiento de Datos

Normalizar las columnas de calificaciones y horas de estudio

```python
from sklearn.preprocessing import StandardScaler

# Normalizar las columnas de calificaciones y horas de estudio
scaler = StandardScaler()
cols = ["Nota_Matemáticas", "Nota_Lenguaje", "Nota_Ciencias", "Horas_de_estudio"]
new_cols = ["Nota_Matemáticas_norm", "Nota_Lenguaje_norm", "Nota_Ciencias_norm", "Horas_de_estudio_norm"]
df[new_cols] = scaler.fit_transform(df[cols])
```

```python
# Mostrar el rango después de la normalización
print("\nRango de las calificaciones después de la normalización:")
df[new_cols].describe()
```

Convertir las categorías de la columna "Genero" en variables numéricas (One-Hot Encoding)

```python
# Convertir 'Genero' en variables numéricas mediante One-Hot Encoding
df_genero_encoded = pd.get_dummies(df['Genero'], prefix='Genero')

# Mostrar las primeras filas del DataFrame con la nueva variable
df_with_encoded_genero = pd.concat([df, df_genero_encoded], axis=1)
print("\nPrimeras filas con la columna 'Genero' codificada:")
df_with_encoded_genero.head()# Convertir 'Genero' en variables numéricas mediante One-Hot Encoding
df_genero_encoded = pd.get_dummies(df['Genero'], prefix='Genero')

# Mostrar las primeras filas del DataFrame con la nueva variable
df_with_encoded_genero = pd.concat([df, df_genero_encoded], axis=1)
print("\nPrimeras filas con la columna 'Genero' codificada:")
df_with_encoded_genero.head()
```

### 7. Visualizaciones

Distribución de calificaciones por asignatura

```python
import matplotlib.pyplot as plt

# Crear figura y ejes
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

# Histograma para cada asignatura
df["Nota_Matemáticas"].plot(kind="hist", bins=50, alpha=0.7, ax=axes[0], color="blue")
axes[0].set_title("Distribución de Calificaciones - Matemáticas")
axes[0].set_xlabel("Calificación")
axes[0].set_ylabel("Frecuencia")

df["Nota_Lenguaje"].plot(kind="hist", bins=50, alpha=0.7, ax=axes[1], color="red")
axes[1].set_title("Distribución de Calificaciones - Lenguaje")
axes[1].set_xlabel("Calificación")
axes[1].set_ylabel("Frecuencia")

df["Nota_Ciencias"].plot(kind="hist", bins=50, alpha=0.7, ax=axes[2], color="green")
axes[2].set_title("Distribución de Calificaciones - Ciencias")
axes[2].set_xlabel("Calificación")
axes[2].set_ylabel("Frecuencia")

# Ajuste del espaciado
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

Promedio de calificaciones por comuna (gráfico de barras)

```python
# Promedio de calificaciones por comuna
comuna_promedio = df.groupby("Comuna")[["Nota_Matemáticas", "Nota_Lenguaje", "Nota_Ciencias"]].mean()

# Crear gráfico de barras para el promedio de calificaciones por comuna con colores personalizados
comuna_promedio.plot(kind="bar", figsize=(12, 6), color=["blue", "red", "green"])

# Personalizar el gráfico
plt.title("Promedio de Calificaciones por Comuna")
plt.xlabel("Comuna")
plt.ylabel("Promedio de Calificación")
plt.xticks(rotation=90)
plt.legend(["Matemáticas", "Lenguaje", "Ciencias"])

# Mostrar gráfico
plt.show()
```

Gráfico de dispersión entre Horas de Estudio y el Promedio
